## Create a new Anaconda environment in Jupyter
1. Log into Scholar's Jupyter hub.
2. Click on the `conda` tab for Anaconda environment setup.
![condaEnv](img/conda_env.png)
3. Click on the `+` sign to create a new Anaconda environment.
![condaNewEnv](img/conda_newenv.png)
4. Give your new environment a name. Make sure the type is in `Python 3`. This will ask Jupyter to initialize a environment that uses `Python 3`.
![condaNewEnvName](img/conda_newenv_name.png)
5. You might need to wait a little bit for Jupyter to initialize the environment. If everything goes right, you will have your new environment listed.
![condaMyEnv](img/conda_myenv.png)
**Sometimes you will receive a timeout error from Jupyter. It is normally not an issue. You can refresh your browser to check if your environment is listed.**

Now, we want to install some custom libraries into our new environment.

## Installations of libraries
1. Log into Scholar's remote desktop using either `ThinLinc` or a web browser.
2. Open up a terminal and do:
```shell
module spider anaconda
```
This should give you all the available `anaconda` versions available on Scholar.
3. What we want is the `anaconda/5.3.1-py37` version. Hence, we will do:
```shell
module load anaconda/5.3.1-py37
```
This will set up the proper paths and environment variables for you in that particular terminal.
4. Now in your terminal, do:
```
source activate myenv
```
**You should replace `myenv` with the name of newly added Anaconda environment.**
This will drop into the Anaconda environment you just added. If it is successful, you should see your environment's name prepended in front of your username in parenthesis.
5. Now we want to install a few libraries, do:
```shell
conda install -n myenv plotly numpy pandas
```
**You should replace `myenv` with the name of newly added Anaconda environment.**
This will install `plotly`, `numpy`, and `pandas` into your Anaconda environment.

## Signup for `plotly` and `Mapbox`
This section is rather straightforward.

You will need to provide your email to sign up for `plotly` and `Mapbox`.

Once you have your accounts for both sites, you can get a correspondent API key from `plotly` and a token from `Mapbox`.

For `plotly`:
1. Go to your account.
2. Click on `API keys`.
![condaPlotly](img/conda_plotly.png)
3. Click on `Regenerate Key`.
4. `plotly` will generate an API key for you. You should save the key in somewhere safe and easy to remember.

For `Mapbox`:
1. Go to your account.
2. Click on `Tokens`.
3. You should see a `default public token` listed.
![condaMapbox](img/conda_mapbox.png)
4. Save your token to somewhere safe and easy to remember.
